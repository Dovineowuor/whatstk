# Weekly Activity Tutorial
In this tutorial we will show how easy it is to get a week activity grid of a WhatsApp group chat using `whatstk` library.

```python
from __future__ import print_function
import sys
sys.path.append('../')
```

```python
from whatstk.wparser import WhatsAppChat
```

As always, we start by creating the `WhatsAppChat` object with the chat file that we want to analyse.

```python
wpchat = WhatsAppChat("../chats/samplechat.txt")
```

```python
df = wpchat.week_hour_grid();
```

```python
from whatstk import wpvis
```

```python
# Set figure size
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
wpvis.vis_week_hour_grid(df, title_size=25, xlabel_size=20, ylabel_size=20, tick_size=15, anot_size=15)
```
